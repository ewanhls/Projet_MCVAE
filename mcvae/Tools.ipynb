```python
#!/usr/bin/env python3

import pandas as pd

def reformat_table(input_path, output_path):
    # Lecture en considérant la première colonne comme index
    df = pd.read_csv(input_path, sep='\t', index_col=0)

    # Remettre l'index en colonne
    df = df.reset_index()
    # df.columns[0] = ancien index, df.columns[1] = noms des patients

    # Echanger l'ancien index et la colonne des patients
    cols = list(df.columns)
    if len(cols) < 2:
        raise ValueError("Le fichier doit contenir au moins deux colonnes (index + 1 colonne).")
    cols[0], cols[1] = cols[1], cols[0]

    # Réordonner le dataframe
    df = df[cols]

    # Supprimer la deuxième colonne (ancien index)
    df = df.drop(df.columns[1], axis=1)

    # Supprimer la dernière colonne
    df = df.iloc[:, :-1]

    # Sauvegarder le fichier
    df.to_csv(output_path, sep='\t', index=False)

    print(f"Fichier traité et sauvegardé dans : {output_path}")




if __name__ == "__main__":
    input_file = "/data/projets/bio-int/Propre/Projet_MCVAE/Data_CPTAC/FeaturesCommunes/normal_transcriptomics_features_communes.tsv"     # <-- à modifier
    output_file = "/data/projets/bio-int/Propre/Projet_MCVAE/Data_CPTAC/FeaturesCommunesPr/normal_transcriptomics_features_communes.tsv"   # <-- à modifier

    reformat_table(input_file, output_file)
```

```python
import pandas as pd

# -----------------------------
# 1. Charger les fichiers TSV
# -----------------------------
file1 = "/data/projets/bio-int/Propre/Projet_MCVAE/Data_CPTAC/FeaturesCommunes/normal_proteomics_features_communes.tsv"
file2 = "/data/projets/bio-int/Propre/Projet_MCVAE/Data_CPTAC/FeaturesCommunes/Ccrcc_proteomics_features_communes.tsv"

df1 = pd.read_csv(file1, sep='\t', index_col=0)
df2 = pd.read_csv(file2, sep='\t', index_col=0)

# -----------------------------
# 2. Extraire les gènes (colonnes)
# -----------------------------
genes1 = set(df1.columns)
genes2 = set(df2.columns)

# -----------------------------
# 3. Stats et comparaison
# -----------------------------
print("Nombre de gènes dans le premier fichier :", len(genes1))
print("Nombre de gènes dans le second fichier :", len(genes2))

common_genes = genes1.intersection(genes2)
print("Nombre de gènes communs :", len(common_genes))

only_in_1 = genes1 - genes2
only_in_2 = genes2 - genes1
print("Nombre de gènes uniquement dans le premier fichier :", len(only_in_1))
print("Nombre de gènes uniquement dans le second fichier :", len(only_in_2))

# -----------------------------
# 4. Affichage des résultats
# -----------------------------
print("\nQuelques gènes communs :", list(common_genes)[:10])
print("\nQuelques gènes uniques au premier fichier :", list(only_in_1)[:10])
print("\nQuelques gènes uniques au second fichier :", list(only_in_2)[:10])

# -----------------------------
# 5. Optionnel : sauvegarder la liste des gènes communs
# -----------------------------
# common_genes_file = "/chemin/vers/common_genes.txt"
# with open(common_genes_file, "w") as f:
#     for g in sorted(common_genes):
#         f.write(g + "\n")

# print(f"\nListe des gènes communs sauvegardée dans : {common_genes_file}")
```

```python
import numpy as np
import pandas as pd

rna = pd.read_csv(
    "/data/projets/bio-int/Propre/Projet_MCVAE/Data_CPTAC/Training/CCRCC/Data/normal_transcriptomics_features_communes.tsv",
    sep='\t',
    index_col=0
)
rppa = pd.read_csv(
    "/data/projets/bio-int/Propre/Projet_MCVAE/Data_CPTAC/Training/CCRCC/Data/Ccrcc_transcriptomics_features_communes.tsv",
    sep='\t',
    index_col=0
)

# Si les colonnes d'ID parasites existent
for col in ['Original_ID', rna.columns[-1], rppa.columns[-1]]:
    if col in rna.columns:
        rna = rna.drop(columns=[col])
    if col in rppa.columns:
        rppa = rppa.drop(columns=[col])

# Vérification des shapes
print("RNA shape:", rna.shape)
print("RPPA shape:", rppa.shape)

# Assurez-vous que toutes les colonnes sont numériques
rna = rna.apply(pd.to_numeric, errors='coerce')
rppa = rppa.apply(pd.to_numeric, errors='coerce')

print(rna.head())
print(rna.dtypes)
print(rna.values.dtype)

# Vérification que les échantillons correspondent
assert all(rna.index == rppa.index), "Les lignes doivent représenter les mêmes échantillons !"



print("\n=== Vérification NaN ===")
print(f"RNA NaN count: {np.isnan(rna.values).sum()}")
print(f"RPPA NaN count: {np.isnan(rppa.values).sum()}")
```
