```python
import pandas as pd
from pathlib import Path

def clean_index_file(path: Path):
    """
    Lit un fichier CSV/TSV, corrige l'index si nÃ©cessaire et sauvegarde le fichier.
    """
    print(f"\nðŸ“„ Processing: {path.name}")

    # DÃ©tecte le sÃ©parateur
    sep = "\t" if path.suffix.lower() == ".tsv" else ","

    # Lecture
    df = pd.read_csv(path, sep=sep)

    # Corrige l'index si une colonne 'Unnamed: 0' apparaÃ®t
    if "Unnamed: 0" in df.columns:
        print("  âž¤ Fixing index column: 'Unnamed: 0'")
        df = df.set_index("Unnamed: 0")

    # Renommer l'index si vide
    if df.index.name in [None, ""]:
        df.index.name = "ID"

    # Sauvegarde
    df.to_csv(path, sep=sep)
    print(f"  ðŸ’¾ Saved cleaned file: {path.name}")


def process_all_files(folder_path: str):
    """
    Parcourt tous les fichiers CSV/TSV dans un dossier et nettoie l'index.
    """
    folder = Path(folder_path)
    if not folder.exists():
        raise FileNotFoundError(f"Dossier introuvable : {folder}")

    print(f"\nðŸš€ Processing all omics datasets in folder: {folder}\n")

    for file in folder.iterdir():
        if file.suffix.lower() in [".csv", ".tsv"]:
            clean_index_file(file)

    print("\nâœ¨ All datasets processed successfully!\n")


# ===================================================
# USAGE
# ===================================================
process_all_files("/data/projets/bio-int/Propre/Projet_MCVAE/Data_CPTAC/FeaturesCommunes")
```
